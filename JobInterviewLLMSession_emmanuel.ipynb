**Which statement correctly identifies the variables?**

- **B)** House price is the independent variable; all three features are dependent variables
- **C)** House price is the dependent variable; square footage, bedrooms, and crime rate are independent variables

### Question 3 of 15 — Train / Validation / Test Split

A junior data scientist splits their data into training and test sets. They use the **test set** to tune hyperparameters (like choosing the best k for KNN), then report the test set accuracy as their final model performance.

**What is the problem with this approach?**

- **A)** There is no problem — this is standard practice
- **B)** The training set should always be smaller than the test set
- **C)** The model will underfit because the test set is too small
- **D)** The test set performance is now an optimistic (biased) estimate because it was used for model selection — a separate validation set should have been used for tuning
- **E)** The problem is that they should have used cross-validation instead of any split at all

**Emmanuel's Answer:** A

**❌ INCORRECT. Correct answer: D.**

**Explanation:** If you use the test set to make decisions (like picking the best k), the model has been indirectly optimized for that test data. When you then report test accuracy as your final score, it's overly optimistic — it no longer reflects how the model would perform on truly unseen data.

**The correct workflow is a three-way split:**
- **Training set** → fit the model
- **Validation set** → tune hyperparameters, compare models
- **Test set** → touch it once, only at the very end, to get an honest final performance estimate

The test set is like a sealed exam. If you practice using the actual exam questions to prepare, your score doesn't reflect real knowledge. The validation set is your practice exam.

### Question 4 of 15 — Linear Regression: Residuals

After fitting a linear regression model, you plot **residuals (y - ŷ) vs. predicted values (ŷ)** and observe a clear **fan shape** — residuals are small for low predictions but spread widely for high predictions.

**What does this pattern indicate?**

- **A)** The model is perfectly fitted with no issues
- **B)** The model suffers from multicollinearity among features
- **C)** The data exhibits **heteroscedasticity** — non-constant variance of errors — violating a linear regression assumption
- **D)** The model is underfitting and needs more features
- **E)** The residuals confirm that the relationship is perfectly linear

**Emmanuel's Answer:** C

**✅ CORRECT.** The fan shape is a classic heteroscedasticity signature — the variance of errors changes across the range of predictions. Linear regression assumes homoscedasticity (constant error variance). Common remedies include log-transforming the target variable, using weighted least squares, or applying other linearization techniques.

### Question 5 of 15 — Regression Analysis: R² and MSE

Two models are trained on the same dataset to predict employee salaries:

- **Model A:** R² = 0.92, MSE = 15,000
- **Model B:** R² = 0.87, MSE = 22,000

**Which interpretation is correct?**

- **A)** Model B is better because its MSE is higher, meaning it captures more variance
- **B)** Model A explains 92% of the variance in salary and has lower average squared error — it is the stronger model on this data
- **C)** R² and MSE measure completely unrelated things, so the two models cannot be compared
- **D)** Model A is overfitting because no model should have an R² above 0.90
- **E)** MSE is only valid for classification tasks, so only R² should be used here

**Emmanuel's Answer:** B — High R² = lower MSE

**✅ CORRECT.** R² and MSE are inversely related on the same dataset. R² tells you the proportion of variance explained (closer to 1.0 = better), while MSE tells you the magnitude of prediction error (lower = better).

### Question 6 of 15 — Logistic Regression: Intercept and Slope

A logistic regression model trained to predict spam emails produces:

- **Intercept (β₀) = -3.5**
- **Coefficient for "exclamation_count" (β₁) = 1.8**

**What is the correct interpretation?**

- **A)** When exclamation_count is 0, there is a 3.5% probability the email is spam
- **B)** Each additional exclamation mark increases the probability of spam by exactly 1.8%
- **C)** When exclamation_count is 0, the log-odds of spam are -3.5, and each additional exclamation mark increases the log-odds by 1.8
- **D)** The intercept means the model is biased and should be removed
- **E)** The coefficient 1.8 means the model is 1.8 times more accurate for emails with exclamation marks

**Emmanuel's Answer:** E

**❌ INCORRECT. Correct answer: C.**

**Explanation:** In logistic regression, the model computes log-odds, not probabilities directly:

**log-odds = β₀ + β₁x = -3.5 + 1.8 × exclamation_count**

The coefficients operate in **log-odds space**, not probability space and not accuracy space.

- **Intercept (β₀ = -3.5):** When exclamation_count = 0, the log-odds of spam are -3.5. Converting to probability: P = 1/(1 + e^3.5) ≈ 0.03, so about 3% chance of spam.
- **Coefficient (β₁ = 1.8):** Each additional exclamation mark adds 1.8 to the log-odds. It does NOT add 1.8 to the probability. The odds ratio is e^1.8 ≈ 6.05, meaning each exclamation mark multiplies the odds of spam by about 6x.

### Question 7 of 15 — Cross-Entropy (Log Loss)

A logistic regression model predicts a **0.95 probability** that an email is spam, and the true label is **spam (1)**. Another email gets a **0.55 probability** of spam, and it is also truly **spam (1)**.

**Which statement about cross-entropy loss is correct?**

- **A)** Both predictions produce the same loss since both are classified correctly at a 0.5 threshold
- **B)** The 0.55 prediction produces lower loss because it is closer to the decision boundary
- **C)** The 0.95 prediction produces lower loss because the model was more confident and correct — log loss penalizes low-confidence correct predictions more heavily
- **D)** Cross-entropy only cares about whether the classification is right or wrong, not the probability
- **E)** Cross-entropy is only used for multi-class problems, not binary classification

**Emmanuel's Answer:** C

**✅ CORRECT.** Cross-entropy loss is calculated as -log(predicted probability for the true class). For the 0.95 prediction: -log(0.95) ≈ 0.05 (very small loss). For the 0.55 prediction: -log(0.55) ≈ 0.60 (much higher loss). Log loss rewards confident, correct predictions and heavily penalizes confident wrong ones.

### Question 8 of 15 — KNN Hyperparameters

You're tuning a KNN classifier and testing k values from 1 to 50. You observe that:

- At **k=1**, training accuracy is 100% but test accuracy is 72%
- At **k=25**, training accuracy is 85% and test accuracy is 88%
- At **k=50**, training accuracy is 70% and test accuracy is 68%

**What best explains this pattern?**

- **A)** k=1 is the best model because it achieves perfect training accuracy
- **B)** k=1 overfits (memorizes training data, poor generalization), k=50 underfits (too much smoothing, loses local patterns), and k=25 balances bias and variance with the best test performance
- **C)** Increasing k always improves performance, so k=50 should be selected
- **D)** KNN is not affected by the choice of k — all three models are equivalent
- **E)** k=50 is best because lower training accuracy proves the model is not overfitting

**Emmanuel's Answer:** B

**✅ CORRECT.** This is the classic bias-variance tradeoff in action. k=1 creates a highly complex, jagged decision boundary that memorizes noise (high variance). k=50 over-smooths everything (high bias). k=25 hits the sweet spot.

### Question 9 of 15 — Decision Trees: Leaf Nodes and Predictions

A decision tree classifier is trained to predict whether a customer will churn (yes/no). One **leaf node** contains 45 samples: 38 labeled "Yes" (churn) and 7 labeled "No."

**How does the tree make a prediction for a new customer that lands in this leaf?**

- **A)** It randomly selects "Yes" or "No" for each new prediction
- **B)** It predicts "No" because the tree always favors the minority class
- **C)** It predicts "Yes" (the majority class in that leaf), and can also report a probability of 38/45 ≈ 84.4%
- **D)** It cannot make a prediction because the leaf is not pure (not 100% one class)
- **E)** It sends the sample back up the tree for reclassification at the root

**Emmanuel's Answer:** C — It has been trained majorly on the leaf node originally, considering most labels said Yes, when a new customer lands in the leaf it will assign it that because it was not there during the original training.

**✅ CORRECT.** The leaf node is a "bucket" — during training, samples trickle down through decision rules. At prediction time, a new sample follows the same path and gets the majority vote. The tree can also output the class distribution (38/45 ≈ 84.4%) as a probability estimate.

### Question 10 of 15 — Decision Trees: Overfitting and Max Depth

You train two decision trees on the same dataset:

- **Tree A:** max_depth=None (unlimited), achieves 99% training accuracy and 71% test accuracy
- **Tree B:** max_depth=3, achieves 84% training accuracy and 83% test accuracy

**What is the best interpretation?**

- **A)** Tree A is better because it has higher training accuracy
- **B)** Tree A with unlimited depth has overfit — it memorized training data by creating overly specific rules. Tree B's depth limit acts as regularization, producing a simpler model that generalizes better
- **C)** Decision trees cannot overfit regardless of depth
- **D)** Tree B is underfitting because its training accuracy is below 90%
- **E)** The depth parameter has no meaningful effect on decision tree performance

**Emmanuel's Answer:** B — That's classic overfitting. When you limit depth, you're basically applying regularization.

**✅ CORRECT.** Unlimited depth lets the tree keep splitting until leaves are nearly pure, creating rules so specific they capture noise. Setting max_depth is one of several regularization techniques for trees — others include min_samples_split, min_samples_leaf, and max_leaf_nodes.

### Question 11 of 15 — Parametric vs. Non-Parametric Models

**Which of the following correctly distinguishes parametric from non-parametric models?**

- **A)** Parametric models have no parameters; non-parametric models have many parameters
- **B)** Parametric models assume a fixed functional form (e.g., linear equation) and learn a fixed number of parameters from data. Non-parametric models make fewer assumptions about the data's structure and can grow in complexity with more data
- **C)** Non-parametric models are always more accurate than parametric models
- **D)** Linear regression is non-parametric because you can add as many features as you want
- **E)** KNN is parametric because you must set the k parameter before training

**Emmanuel's Answer:** B — Key words "fixed" and "grow"

**✅ CORRECT.** Linear regression is parametric — no matter how much data you throw at it, it learns a fixed set of coefficients. KNN is non-parametric — despite having the hyperparameter k, it stores the entire training dataset and its effective complexity grows with data size. Option E is a common trap: k is a hyperparameter you set, not a learned parameter.

### Question 12 of 15 — Data Leakage (Scenario-Based)

You're building a model to predict whether hospital patients will develop sepsis within 48 hours of admission. Your dataset includes a feature called `antibiotic_prescribed` — whether the patient was given a specific antibiotic.

Your model achieves 97% accuracy during testing. Your manager is thrilled, but a senior data scientist raises a concern.

**What is the most likely problem?**

- **A)** The dataset is too large, causing the model to overfit
- **B)** 97% accuracy is always suspicious — the model must be broken
- **C)** The `antibiotic_prescribed` feature likely leaks future information — doctors prescribe that antibiotic because they already suspect sepsis, so the feature encodes the outcome rather than predicting it
- **D)** The model should use a deeper decision tree to be more trustworthy
- **E)** Antibiotics have no relationship to sepsis, so the feature is just noise that got lucky

**Emmanuel's Answer:** C — Because of the fear that feature is likely to leak future information

**✅ CORRECT.** This is a textbook target leakage scenario. The antibiotic wasn't prescribed at admission — it was prescribed after doctors suspected sepsis. In production, at the moment of admission, you wouldn't have that feature yet. Rule of thumb: For every feature, ask — "Would I have this information at the exact moment I need to make the prediction?"

### Question 13 of 15 — Imbalanced Data (Scenario-Based)

You're building a fraud detection model. Only **0.5%** of transactions are fraudulent. You train a classifier and it achieves **99.5% accuracy**.

**What should your reaction be?**

- **A)** The model is excellent — 99.5% accuracy is outstanding
- **B)** The model has likely learned to predict "not fraud" for everything — matching the class distribution. Accuracy is misleading here; you should evaluate using precision, recall, F1-score, or the precision-recall curve instead
- **C)** You should remove the majority class to balance the dataset to exactly 50/50
- **D)** Accuracy above 99% always indicates overfitting
- **E)** The model needs more features to push accuracy above 99.9%

**Emmanuel's Answer:** B — This is a severe class imbalance problem. Accuracy becomes misleading because it is dominated by the majority class.

**✅ CORRECT.** A dumb model that just prints "not fraud" every single time would score 99.5% accuracy. In imbalanced scenarios, precision tells you "of the transactions I flagged as fraud, how many actually were?", recall tells you "of all actual fraud, how many did I catch?", and F1 balances both.
