- **C.** Supervised learning, because customer data always has a target variable.

**Your Answer:** B

**✅ CORRECT.**

### Question 2 of 15 — Dependent vs. Independent Variables

In a linear regression model predicting house prices, you have: square footage, number of bedrooms, proximity to schools, and **sale price**.

Which of these is the **dependent variable**?

- **A.** Square footage; it is the variable we manipulate to control the output.
- **B.** Sale price; it is the variable the model is trying to predict, and it depends on the other features.
- **C.** Number of bedrooms; it has the strongest correlation with sale price.
- **D.** Proximity to schools; it is the most complex variable to measure.
- **E.** All four variables are dependent because they all influence each other.

**Your Answer:** B

**✅ CORRECT.**

**Explanation:** The dependent variable (also called the target or label) is what the model is trying to predict — it 'depends' on the independent variables (features) you feed in.

### Question 3 of 15 — Train / Validation / Test Split & Data Leakage

A junior data scientist builds a model to predict loan defaults. She splits her data into training and test sets only, tunes the model's hyperparameters repeatedly against the test set until she gets the best accuracy, then reports that accuracy as her final result.

What is the fundamental problem?

- **A.** She should have used more features to improve the model.
- **B.** The test set has been implicitly used for training decisions, so it no longer provides an unbiased estimate of real-world performance — this is a form of data leakage.
- **C.** Hyperparameter tuning should only be done on the training set using a single fixed configuration.
- **D.** Loan default prediction always requires a 3-way split regardless of dataset size.
- **E.** There is no problem — repeatedly testing against the test set is standard practice.

**Your Answer:** B

**✅ CORRECT.**

**Explanation:** Once you make decisions based on test set performance, it becomes contaminated — your reported accuracy is optimistic and won't generalize. The correct approach: train/validation/test split — tune on the validation set, touch the test set exactly once at the end.

### Question 4 of 15 — MSE & R² Interpretation

After fitting a linear regression model: MSE = 452.3, R² = 0.91. A colleague says MSE is too high — the model is useless. Another says R² is great.

Who is more correct?

- **A.** The first colleague is right — MSE above 100 always indicates a poor model.
- **B.** The second colleague is right — R² is always the more important metric and MSE can be ignored.
- **C.** Both metrics must be considered together. R² = 0.91 means the model explains 91% of the variance, which is strong. MSE must be evaluated relative to the scale of the target variable.
- **D.** Neither metric is useful for linear regression; you should only use accuracy.
- **E.** MSE = 452.3 means the model is wrong 452 times out of every 1000 predictions.

**Your Answer:** C

**✅ CORRECT.**

**Explanation:** MSE is in squared units of the target — if you're predicting house prices in dollars, an MSE of 452 might mean an average error of ~$21, which is excellent. You can never judge MSE in isolation.

### Question 5 of 15 — Logistic Regression: Intercept & Slope (Log-Odds)

A logistic regression spam classifier has learned:
- **Intercept (bias):** -4.2
- **Coefficient for "number of exclamation marks":** 1.8

What is the correct interpretation of the coefficient 1.8?

- **A.** For every additional exclamation mark, the probability of spam increases by 1.8%.
- **B.** For every additional exclamation mark, the **log-odds** of the email being spam increase by 1.8.
- **C.** The model will classify any email with more than 1.8 exclamation marks as spam.
- **D.** 1.8 is the probability threshold above which the email is labeled spam.
- **E.** The coefficient means exclamation marks reduce the spam probability since 1.8 is less than 2.

**Your Answer:** D

**❌ INCORRECT. Correct answer: B.**

**Explanation:** In logistic regression, coefficients operate in **log-odds space**, not probability space. A coefficient of 1.8 means that for each additional exclamation mark, the log-odds of being spam increase by 1.8. That increase is passed through the **sigmoid function** to produce a probability — but the relationship is nonlinear.

- **A** is wrong — the coefficient doesn't directly map to a percentage probability change.
- **C** is wrong — there's no hard threshold baked into the coefficient.
- **D** is wrong — the decision threshold (commonly 0.5) is a separate concept from learned coefficients.
- **E** is wrong — a positive coefficient always pushes toward the positive class.

The log-odds interpretation is a frequent interview question. Key formula: `log(p/(1-p)) = β₀ + β₁x₁ + ...`

### Question 6 of 15 — Cross-Entropy Loss

Your logistic regression model outputs a **cross-entropy loss of 0.08** on the validation set. A second model produces **0.63**.

What does this tell you, and which model would you prefer?

- **A.** The second model is better because higher loss means more learning has occurred.
- **B.** The first model is better — a lower cross-entropy loss means its predicted probabilities are closer to the true labels.
- **C.** Cross-entropy loss cannot be used to compare two models — you need accuracy instead.
- **D.** Both models are equivalent; cross-entropy only matters during training, not evaluation.
- **E.** The first model is overfit because a loss of 0.08 is suspiciously low.

**Your Answer:** B

**✅ CORRECT.**

**Explanation:** Lower cross-entropy loss means the model's predicted probabilities are consistently close to the true labels. A loss of 0.08 on the validation set is strong and not inherently suspicious — a big gap between training and validation loss would signal overfitting, but a low validation loss on its own is a good sign.

### Question 7 of 15 — KNN Hyperparameter Tuning & Overfitting

KNN classifier for disease prediction:
- **k=1:** Train acc = 100%, Val acc = 61%
- **k=50:** Train acc = 74%, Val acc = 73%
- **k=200:** Train acc = 68%, Val acc = 67%

What is happening at k=1, and what is the best choice?

- **A.** k=1 is the best choice because 100% training accuracy means the model learned perfectly.
- **B.** k=1 is severely overfit. k=50 is the best choice — it generalizes best with the smallest train/val gap.
- **C.** k=200 is best because larger k always produces more stable and accurate models.
- **D.** k=1 is underfit because a single neighbor cannot capture enough complexity.
- **E.** None are valid — k must always be odd in KNN.

**Your Answer:** B

**✅ CORRECT.**

**Explanation:** At k=1, the model memorizes every training point — perfect training accuracy but poor generalization (textbook overfitting). k=50 hits the sweet spot. k=200 starts to underfit — too much smoothing loses meaningful signal.

### Question 8 of 15 — Decision Tree Leaf Nodes

A trained decision tree classifies fruit (Apple, Banana, Orange). A leaf node contains: **8 Apples, 1 Banana, 1 Orange**.

What will this leaf node predict for any new sample that reaches it?

- **A.** It will predict Banana, because it is the minority class and needs more representation.
- **B.** It will predict randomly among the three classes with equal probability.
- **C.** It will predict Apple, because the majority class at a leaf node determines the prediction.
- **D.** It will predict Orange, because decision trees always predict the last class alphabetically.
- **E.** It cannot make a prediction until all samples in the leaf belong to the same class.

**Your Answer:** C

**✅ CORRECT.**

**Explanation:** Leaf nodes use majority voting — whatever class dominates the training samples that landed in that node becomes the prediction for any new sample routed there. The 'impurity' (Gini/entropy) is measured during training, but at prediction time it's simply the majority class.

### Question 9 of 15 — Parametric vs. Non-Parametric Trade-offs

A colleague argues: 'Linear regression is parametric, KNN is non-parametric. That means KNN is always more flexible and therefore better.'

What is wrong with this reasoning?

- **A.** Nothing — non-parametric models are universally superior to parametric ones.
- **B.** The colleague has the definitions backwards — KNN is parametric and linear regression is non-parametric.
- **C.** The definitions are correct, but 'more flexible' doesn't mean 'better.' KNN's flexibility comes with costs: computationally expensive at prediction time, sensitive to irrelevant features and scale, and can overfit with small k.
- **D.** KNN is non-parametric only when k=1; for all other values it becomes parametric.
- **E.** Parametric vs. non-parametric is an outdated distinction that no longer applies.

**Your Answer:** C

**✅ CORRECT.**

**Explanation:** Flexibility is not free — it comes with computational cost (KNN must store and search all training data at inference), sensitivity to feature scale (requiring normalization), and susceptibility to the curse of dimensionality. Always frame model selection as a trade-off, not a ranking.

### Question 10 of 15 — Imbalanced Data & Accuracy Pitfall

Fraud detection dataset: **98% legitimate**, **2% fraudulent**. Model achieves **98% accuracy**. Stakeholder is thrilled. Should you be?

- **A.** Yes — 98% accuracy is excellent by any standard and the model is ready for deployment.
- **B.** No — a model predicting 'legitimate' for every transaction also achieves 98% accuracy. Accuracy is misleading on imbalanced datasets; use precision, recall, and F1, focusing especially on recall for the fraud class.
- **C.** No — accuracy above 95% always indicates overfitting on imbalanced data.
- **D.** Yes, but only if the model was trained on a balanced dataset first.
- **E.** It depends on whether the model used logistic regression or a decision tree.

**Your Answer:** B

**✅ CORRECT.**

**Explanation:** A naive majority-class classifier achieves 98% accuracy here while completely failing at its actual job — catching fraud. In imbalanced settings, recall (did we catch the fraud?) and precision (of those we flagged, how many were actually fraud?) matter far more than raw accuracy.

### Question 11 of 15 — Linearization of Nonlinear Relationships

Your data suggests y grows **exponentially** with x. Which transformation best linearizes this relationship?

- **A.** Square x — regress y on x².
- **B.** Take the natural log of y — regress log(y) on x, linearizing the exponential into log(y) = mx + b.
- **C.** Take the square root of x — regress y on √x.
- **D.** Standardize both variables to mean 0 and standard deviation 1 before regressing.
- **E.** Exponential relationships cannot be linearized — you must use a nonlinear model.

**Your Answer:** B

**✅ CORRECT.**

**Explanation:** If y = e^(mx+b), then log(y) = mx + b — a clean linear relationship. This is the core idea behind linearization: transform variables so the relationship becomes linear, fit the model, then back-transform if needed.

### Question 12 of 15 — Temporal Data Leakage & Concept Drift

Hospital readmission model trained on 2018–2022 data. You randomly split 80/20, get strong results. Deployed 2023 — performance is significantly worse.

Most likely cause?

- **A.** The model was not complex enough — you should have used a deeper neural network.
- **B.** Random splitting of temporal data causes leakage — future data leaks into training. Patient patterns also shift (concept drift). You should have used a temporal split: train on 2018–2021, test on 2022.
- **C.** The test set was too small at 20%.
- **D.** Hospital data always requires cross-validation with at least 10 folds.
- **E.** The model overfit because medical data has too many features.

**Your Answer:** B

**✅ CORRECT.**

**Explanation:** Two issues: (1) temporal leakage — randomly shuffling time-ordered data lets the model train on 'future' records; (2) concept drift — patient demographics and treatment protocols shift over time. Always respect time order when splitting temporal data.

### Question 13 of 15 — KNN Hyperparameters vs. Learned Parameters

Which of the following is **not** a hyperparameter in the KNN algorithm?

- **A.** The value of k (number of neighbors).
- **B.** The distance metric (e.g., Euclidean, Manhattan).
- **C.** The weighting scheme (uniform vs. distance-weighted votes).
- **D.** The learned feature weights assigned during training.
- **E.** The algorithm used to compute nearest neighbors (e.g., ball tree, KD-tree).

**Your Answer:** D

**✅ CORRECT.**

**Explanation:** KNN has no learned parameters at all — it's a lazy learner that memorizes training data and defers all computation to prediction time. There are no weights learned during training, unlike in linear regression or neural networks. This is precisely what makes KNN non-parametric.

### Question 15 of 15 — Multi-constraint Model Deployment Decision

Real-time fraud detection API (must respond in <50ms; missing fraud is more costly than false alarm):

| Metric | Logistic Regression | KNN (k=9) |
|---|---|---|
| Accuracy | 86% | 86% |
| Precision | 0.81 | 0.74 |
| Recall | 0.72 | 0.84 |
| Prediction time | 1ms | 890ms |

Which model should you deploy and why?

- **A.** KNN — higher recall means it catches more fraud, and that's all that matters.
- **B.** Logistic Regression — faster prediction time meets the 50ms SLA, and while recall is lower, the difference is manageable.
- **C.** Logistic Regression — but lower the decision threshold below 0.5 to increase recall, meeting both the latency requirement and better aligning with the cost of missing fraud.
- **D.** KNN — prediction time can be optimized later with approximate nearest neighbor libraries.
- **E.** Neither — you need a more powerful model like a neural network for fraud detection.

**Your Answer:** C

**✅ CORRECT.**

**Explanation:** This is the best possible answer. Two constraints identified simultaneously: (1) **Latency** — KNN at 890ms catastrophically fails the 50ms SLA and is disqualified regardless of recall advantage; (2) **Business cost asymmetry** — missing fraud is worse than a false alarm, so lower the threshold below 0.5 to get higher recall for free without retraining.
