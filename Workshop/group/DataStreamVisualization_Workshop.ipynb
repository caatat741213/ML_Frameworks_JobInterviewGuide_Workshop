## 🌞 StreamingSimulator

```python
import pandas as pd
import time
from sqlalchemy import create_engine

# Define the StreamingSimulator class
class StreamingSimulator:
    def __init__(self, relative_path):
        """
        Initialize the simulator with a CSV file path.
        """
        try:
            # Load the CSV data into a DataFrame
            self.df = pd.read_csv(relative_path)
            self.current_index = 0
            print(f"✅ Successfully loaded {len(self.df)} records.")
        except FileNotFoundError:
            print(f"❌ Error: File not found at {relative_path}")

    def nextDataPoint(self):
        """
        Load a single record for processing.
        """
        if self.current_index < len(self.df):
            # Get the record at the current index
            data_point = self.df.iloc[[self.current_index]]
            self.current_index += 1
            return data_point
        else:
            return None

# Instantiate the object 'ss' as per workshop requirements
csv_path = "data/RMBR4-2_export_test.csv"
ss = StreamingSimulator(csv_path)

# Test the method: Invoke nextDataPoint
# sample = ss.nextDataPoint()
# if sample is not None:
#     print("First data point loaded:")
#     display(sample)
```

## 🌞 Database & Visualization


```python
import time
import pandas as pd
import psycopg2
from IPython.display import clear_output
import matplotlib.pyplot as plt

class StreamingSimulator:
    def __init__(self, file_path, db_url):
        # Initializing simulator attributes
        self.file_path = file_path
        self.db_url = db_url
        self.df = None
        self.conn = None
        self.cur = None

    def load_data(self):
        # Load CSV and handle missing values & types
        self.df = pd.read_csv(self.file_path)
        
        # Pruning: Remove empty sensor axes (9-14)
        self.df = self.df.drop(columns=[col for col in self.df.columns if self.df[col].isna().all()])
        
        # Ensure 'Time' is in datetime format
        self.df["Time"] = pd.to_datetime(self.df["Time"], utc=True)
        print(f"Data Cleaned. Found {len(self.df)} streamable points.")

    def init_database(self):
        # Connect to Neon.tech and create the table schema
        self.conn = psycopg2.connect(self.db_url)
        self.conn.autocommit = True
        self.cur = self.conn.cursor()

        # Drop old table and recreate to ensure column names match
        self.cur.execute("DROP TABLE IF EXISTS robot_telemetry;")

        # Create table if not exists
        self.cur.execute("""
            CREATE TABLE robot_telemetry (
                id BIGSERIAL PRIMARY KEY,
                time TIMESTAMPTZ,
                trait TEXT,
                axis1 DOUBLE PRECISION, axis2 DOUBLE PRECISION, axis3 DOUBLE PRECISION,
                axis4 DOUBLE PRECISION, axis5 DOUBLE PRECISION, axis6 DOUBLE PRECISION,
                axis7 DOUBLE PRECISION, axis8 DOUBLE PRECISION
            );
        """)

        # Clear old data for a fresh start
        self.cur.execute("TRUNCATE TABLE robot_telemetry;")
        print("Connected to Neon.tech. Table 'robot_telemetry' is ready.")

    def nextDataPoint(self, start_idx=100, num_points=100, wait_time=2):
        """
        start_idx: Starting row (default 0)
        num_points: Number of rows to read (default None means read all)
        wait_time: Simulation interval time (default 0.1 seconds)
        """
        insert_sql = """
        INSERT INTO robot_telemetry (time, trait, axis1, axis2, axis3, axis4, axis5, axis6, axis7, axis8)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        """
        
        # Select data range based on parameters
        if num_points is None:
            stream_data = self.df.iloc[start_idx:]
        else:
            stream_data = self.df.iloc[start_idx : start_idx + num_points]
        
        total = len(stream_data)
        print(f"🚀 Starting stream from index {start_idx}, total points: {total}")

        try:
            for i, (original_idx, row) in enumerate(stream_data.iterrows()):
                # 1. Insert data for 8 axes
                self.cur.execute(insert_sql, (
                    row['Time'].to_pydatetime(), row['Trait'], 
                    row['Axis #1'], row['Axis #2'], row['Axis #3'], row['Axis #4'],
                    row['Axis #5'], row['Axis #6'], row['Axis #7'], row['Axis #8']
                ))

                # 2. Refresh chart every X points (adjust for smoother streaming)
                df_live = pd.read_sql("SELECT * FROM robot_telemetry ORDER BY time DESC LIMIT 50", self.conn)
                df_live = df_live.sort_values("time")

                clear_output(wait=True)
                plt.figure(figsize=(15, 7), facecolor='#f8f9fa')
                
                colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f']
                for j in range(1, 9):
                    col_name = f'axis{j}'
                    lw = 3 if j == 3 else 1
                    alpha = 1.0 if j == 3 else 0.5
                    plt.plot(df_live['time'], df_live[col_name], label=f'Axis #{j}', color=colors[j-1], linewidth=lw, alpha=alpha)
                
                # Draw warning threshold
                # plt.axhline(y=18.0, color='red', linestyle='--', label='Alert (18.0A)')
                
                # Title shows progress
                plt.title(f"Robot Monitor | Progress: {i+1}/{total} | CSV Row: {original_idx}", fontsize=14)
                plt.ylabel("Current (A)")
                plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
                plt.grid(True, alpha=0.3)
                plt.ylim(-1, 40)
                plt.show()

                time.sleep(wait_time)

        except KeyboardInterrupt:
            print("\n⏹️ Stream interrupted by user.")
        except Exception as e:
            print(f"❌ Error: {e}")
        finally:
            if self.cur: self.cur.close()
            if self.conn: self.conn.close()
            print("✅ Connection closed.")

# Execution
DB_URI = "postgresql://neondb_owner:npg_7oErb0uJxlTU@ep-weathered-moon-ah4afhfs-pooler.c-3.us-east-1.aws.neon.tech/neondb?sslmode=require"

# Instantiate as 'ss' per instruction
ss = StreamingSimulator("data/RMBR4-2_export_test.csv", DB_URI)
ss.load_data()
ss.init_database()

# Start the simulation
ss.nextDataPoint()
```

## 🌞Cycle Analysis & Correlation Plot

### Cycle Analysis:
Analyzing a typical 30-second cycle by automatically capturing a continuous motion segment to display the periodic pattern over 30 seconds.

### Axis Correlation Analysis:
The Joint Correlation Map demonstrates how Axis #2 and Axis #3 "collaborate" to influence the robot's health.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# 1. Connect and read data
DB_URI = "postgresql://neondb_owner:npg_7oErb0uJxlTU@ep-weathered-moon-ah4afhfs-pooler.c-3.us-east-1.aws.neon.tech/neondb?sslmode=require"
engine = create_engine(DB_URI)
df = pd.read_sql("SELECT * FROM robot_telemetry ORDER BY time ASC", engine)

# 2. Normalize columns
df.columns = df.columns.str.strip().str.lower()
df['time'] = pd.to_datetime(df['time'])

# Automatically detect all columns containing 'axis'
axis_cols = [col for col in df.columns if 'axis' in col]

if not axis_cols:
    print("❌ No columns containing 'axis' found! Current columns are:", df.columns.tolist())
else:
    # 3. Automatically find the interval with activity (Smart Window Search)
    # Use the first detected axis to find fluctuations
    target_axis = axis_cols[0] 
    df['activity'] = df[target_axis].rolling(window=50).std()
    
    # Find the point with the highest standard deviation (most activity)
    peak_idx = df['activity'].idxmax()
    
    # If no fluctuation is found (e.g., all values are constant), display the first 200 rows
    if pd.isna(peak_idx):
        print("⚠️ No current fluctuations detected (data might be constant). Displaying initial segment.")
        sample_cycle = df.head(200)
    else:
        start_idx = max(0, peak_idx - 100)
        end_idx = min(len(df), peak_idx + 100)
        sample_cycle = df.iloc[start_idx:end_idx]

    # Calculate dynamic threshold based on mean and standard deviation
    mean_current = sample_cycle[axis_cols].mean().mean()
    std_current = sample_cycle[axis_cols].std().mean()
    dynamic_threshold = mean_current + 5 * std_current

    print(f"Calculated dynamic threshold: {dynamic_threshold:.2f}A")

    # --- Plot A: Operational Cycles ---
    plt.figure(figsize=(14, 6))
    for col in axis_cols[:4]: # Plot the first four detected axes
        plt.plot(sample_cycle['time'], sample_cycle[col], label=col.upper(), linewidth=1.5)

    plt.axhline(y=dynamic_threshold, color='red', linestyle='--', alpha=0.6, label=f'Warning ({dynamic_threshold:.2f}A)')
    plt.title("Pattern A: Dynamic Operational Cycles", fontsize=14)
    plt.xlabel("Time")
    plt.ylabel("Current (A)")
    plt.legend()
    plt.grid(True, linestyle=':', alpha=0.5)
    plt.show()

    # --- Plot B: Correlation Matrix ---
    plt.figure(figsize=(10, 8))
    corr_matrix = df[axis_cols].corr()
    plt.imshow(corr_matrix, cmap='RdYlGn', interpolation='none')
    plt.colorbar(label='Correlation')
    plt.xticks(range(len(axis_cols)), [c.upper() for c in axis_cols], rotation=45)
    plt.yticks(range(len(axis_cols)), [c.upper() for c in axis_cols])
    
    for i in range(len(axis_cols)):
        for j in range(len(axis_cols)):
            plt.text(j, i, f'{corr_matrix.iloc[i, j]:.2f}', ha='center', va='center')
            
    plt.title("Pattern B: Axis Synergy Map", fontsize=14)
    plt.show()
```
